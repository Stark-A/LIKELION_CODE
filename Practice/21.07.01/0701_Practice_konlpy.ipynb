```python
from konlpy.tag import Okt
t = Okt()

doc_nouns = t.nouns(doc_ko)
print( len(doc_nouns))
doc_nouns
```

```python
ko = nltk.Text(doc_nouns, name = '크루엘라')
print(type(ko))
print(len(ko.tokens))
```

```python
ko.vocab().most_common(20)
```

```python
most_fre = ko.vocab().most_common(50)
most_fre
```

```python
f = open('stop_words1.txt')
lines = f.readlines()


stop_words1 = []
for i in lines:
  stop_words1.append(i.strip())

new_ko = []
for one_word in ko:
  if one_word not in stop_words1:
    new_ko.append(one_word)
```
