```python
import pandas as pd
```

```python
dat = pd.read_csv('제주특별자치도 서귀포시_유흥주점현황_20200820.csv', engine='python', encoding='EUCKR')
dat
```

```python
dat.columns
```

```python
dat_a = dat.loc[:,['업소명','위도','경도']]
size = dat_a.업소명.count()
print(type(size))

print(dat_a.columns)
```

```python
import folium
from folium import plugins
import os
```

```python
jeju = folium.Map([33.42497006551916, 126.544366564118], zoom_start=10)
for i in range(size):
  folium.Marker(location=[ dat_a.loc[i, '위도'], 
                          dat_a.loc[i, '경도']],
               tooltip=dat_a.loc[i, '업소명'],
                popup=dat_a.loc[i, '업소명'], ).add_to(jeju)
jeju.save('jeju.html')
jeju

```
