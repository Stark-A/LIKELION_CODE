```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

### 실습 1-1
 * sex, day를 그룹화 시켜서 최대값을 출력해 보자.
    
### 실습 1-2
 * sex, day를 그룹화 시켜서 최대값의 time과 size열만 출력해보자.
    
### 실습 1-3
 * 남성의 데이터만 따로 만들어보자. (day가 Sun만 뽑기)male_sun
 * 여성의 데이터만 따로 만들어보자. (day가 Sun만 뽑기)female_sun

```python
g1 = tips.groupby(['sex','day']).max()
g1
```

```python
g2 = tips.groupby(['sex','day'])
g2.max()[['tip','size']]
```

```python
male = tips.groupby(['sex','day']).get_group(('Male','Sun'))
print(male[['sex','day']])
```

```python
female = tips.groupby(['sex','day']).get_group(('Female','Sun'))
print(female[['sex','day']])
```
