```python
doc_ko = open('15_TheExtreme_utf8.txt').read()
doc_ko[1:1000]
```

```python
from konlpy.tag import Okt
t = Okt()

doc_nouns = t.nouns(doc_ko)
print( len(doc_nouns))
doc_nouns
```

```python
ko = nltk.Text(doc_nouns, name = '분노의질주')
print(type(ko))
print(len(ko.tokens))
```

```python
ko.vocab().most_common(20)
```

```python
most_fre = ko.vocab().most_common(50)
most_fre
```

```python
### 한글에서는 따로 불용어 사전이 없어, 따로 만들거나 또는 파일로 부터 불러올 수 있다.
stop_words = ['분노', '영화', '액션', '시리즈', '더', 
              '그', '이', '것', '또', '좀', 
              '돈', '것', '다음', '질주', '그냥', 
              '분노의질주', '말', '뭐', '애', '나', '듯', '편', '볼', '점', '중', '로']
new_ko = [ ]
for one_word in ko:
  if one_word not in stop_words:
    new_ko.append(one_word)
```

```python
### nltk Text 객체 만들기
new_ko = nltk.Text(new_ko, name="분노의 질주2")
plt.figure(figsize=(12,6))
new_ko.plot(50)
```

```python
most_fre = new_ko.vocab().most_common(50)
most_fre
```

```python
f = open('stop_words.txt')
lines = f.readlines()


stop_words = []
for i in lines:
  stop_words.append(i.strip())

new_ko_2 = []
for one_word in new_ko:
  if one_word not in stop_words:
    new_ko_2.append(one_word)
```

```python
### nltk Text 객체 만들기
new_ko2 = nltk.Text(new_ko_2, name='분노의 질주2')
plt.figure(figsize=(12,6))
new_ko2.plot(50)
```
