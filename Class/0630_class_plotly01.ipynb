### 실습
 * x : 인구, y : 기대수명
 * 1행, 2열
 * (px) 산점도 그리기
   * 01 plotly, subplots, make_subplots로 형태 만들기
   * 02 틀안에 들어갈 각각의 그래프 오브젝트 만들기
   * 03 add_trace로 연결시키기
   * (추가) 타이틀 등등을 업데이트 하겠다.
        * fig.update_layout(layout)

```python
fig = make_subplots(rows=1, cols=2, shared_yaxes=True)

gap_1952 = gapminder.query('year==1952')
gap_2007 = gapminder.query('year==2007')

g1952 = go.Scatter(x = gap_1952["pop"], y=gapminder['lifeExp'])
g2007 = go.Scatter(x = gap_2007["pop"], y=gapminder['lifeExp'])

fig.add_trace(g1952, row=1, col=1)
fig.add_trace(g2007, row=1, col=2)

fig
```

```python
layout = {
    "xaxis": {
        "title": "인구수",
    },
    "yaxis": {
        "title": "기대수명"
    },
    "xaxis2": {
        "title": "인구수",
    },
    "yaxis2": {
        "title": "기대수명"
    }
}
fig.update_layout(layout)
```

### 실습 5-1 (추가) - plotly 활용
 * 2행, 2열
 * 1행 - 인구, 기대수명 (1952년, 2007년)
 * 2행 - 인구, 1인당 국민소득(1952년, 2007년)

```python
fig = make_subplots(rows=2, cols=2, shared_xaxes=True)

gap_1952 = gapminder.query('year==1952')
gap_2007 = gapminder.query('year==2007')

g1 = go.Scatter(x = gap_1952['pop'], y = gap_1952['lifeExp'],mode='markers',name='1952년 인구, 기대수명')
g2 = go.Scatter(x = gap_2007['pop'], y = gap_2007['lifeExp'],mode='markers',name='2007년 인구, 기대수명')
g3 = go.Scatter(x = gap_1952['pop'], y = gap_1952['gdpPercap'],mode='markers',name='1952년 인구, GDP')
g4 = go.Scatter(x = gap_2007['pop'], y = gap_2007['gdpPercap'],mode='markers',name='2007년 인구, GDP')

fig.add_trace(g1, row=1, col=1)
fig.add_trace(g2, row=1, col=2)
fig.add_trace(g3, row=2, col=1)
fig.add_trace(g4, row=2, col=2)
layout = {
    "xaxis": {
        "title": "인구수",
    },
    "yaxis": {
        "title": "기대수명"
    },
    "xaxis2": {
        "title": "인구수",
    },
    "yaxis2": {
        "title": "기대수명"
    },
    "xaxis3": {
        "title": "인구수",
    },
    "yaxis3": {
        "title": "1인당 국민소득"
    },
    "xaxis4": {
        "title": "인구수",
    },
    "yaxis4": {
        "title": "1인당 국민소득"
    }
}
fig.update_layout(layout)
```
